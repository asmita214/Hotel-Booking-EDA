• Resort Hotels receive comparatively fewer bookings, suggesting they are more popular for seasonal vacations rather than regular travel.

There is no direct indication of negative growth from this chart. However, the lower booking volume for Resort Hotels highlights an opportunity to improve marketing and customer engagement strategies.
