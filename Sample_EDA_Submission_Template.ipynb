## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df1 = df.copy()
df1.drop_duplicates(inplace=True)
df1['children'] = df1['children'].fillna(df1['children'].median())
df1['country'] = df1['country'].fillna(df1['country'].mode()[0])
df1['agent'] = df1['agent'].fillna(0)
df1['company'] = df1['company'].fillna(0)
print("Dataset Shape :", df1.shape)
print("\nMissing Values:\n")
print(df1.isnull().sum())
````

### What all manipulations have you done and insights you found?

### Data Manipulations Performed

• Loaded the Hotel Booking Demand dataset into a Pandas DataFrame.
• Explored the dataset using head(), shape(), info(), describe(), and columns.
• Identified duplicate records and removed 31,994 duplicate rows.
• Detected missing values in the columns company, agent, country, and children.
• Filled missing values:
   - children → Median
   - country → Mode
   - agent → 0
   - company → 0
• Verified that the cleaned dataset contained no missing values.
• Checked the unique values for every variable to understand categorical and numerical features.

### Key Insights

• Final cleaned dataset contains 87,396 observations and 32 features.
• The dataset includes both categorical and numerical variables, making it suitable for EDA.
• Hotel bookings span three years (2015–2017).
• Variables such as ADR, Lead Time, and Previous Bookings show large variability, indicating possible outliers.
• Binary variables like is_canceled and is_repeated_guest are useful for comparison-based analysis.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
plt.figure(figsize=(8,5))
sns.countplot(
    data=df1,
    x='hotel',
    palette='Set2'
)
plt.title('Distribution of Hotel Bookings by Hotel Type',
          fontsize=16,
          fontweight='bold')
plt.xlabel('Hotel Type',fontsize=12)
plt.ylabel('Number of Bookings',fontsize=12)
plt.grid(axis='y',alpha=0.3)
plt.show()
````

##### 1. Why did you pick the specific chart?

A count plot was chosen because the variable 'hotel' is categorical with only two categories (City Hotel and Resort Hotel). This chart provides a simple and effective way to compare the number of bookings between the two hotel types and helps understand the overall distribution of bookings in the dataset.

##### 2. What is/are the insight(s) found from the chart?

• City Hotels received significantly more bookings than Resort Hotels.

• Approximately 54,000 bookings belong to City Hotels, whereas Resort Hotels account for around 34,000 bookings.

• This indicates that customers prefer City Hotels more frequently, possibly due to business travel, accessibility, and urban tourism.

• Resort Hotels receive comparatively fewer bookings, suggesting they are more popular for seasonal vacations rather than regular travel.


##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. This analysis can help hotel management understand customer preferences and allocate resources accordingly.

Business Impact:
• City Hotels should maintain sufficient room availability and staff to handle higher demand.
• Resort Hotels can improve occupancy by introducing promotional offers, seasonal discounts, family vacation packages, or weekend deals.
• Understanding booking distribution helps optimize pricing, marketing strategies, and operational planning.

There is no direct indication of negative growth from this chart. However, the lower booking volume for Resort Hotels highlights an opportunity to improve marketing and customer engagement strategies.

#### Chart - 2

````python
plt.figure(figsize=(9,6))
sns.countplot(
    data=df1,
    x='hotel',
    hue='is_canceled',
    palette='viridis'
)
plt.title(
    'Hotel Type vs Booking Cancellation',
    fontsize=16,
    fontweight='bold'
)
plt.xlabel('Hotel Type', fontsize=12)
plt.ylabel('Number of Bookings', fontsize=12)
plt.legend(
    title='Cancelled',
    labels=['Not Cancelled', 'Cancelled']
)
plt.grid(axis='y', alpha=0.3)
plt.show()
````

##### 1. Why did you pick the specific chart?

A grouped count plot was selected to compare booking cancellations across the two hotel types. This visualization clearly shows the relationship between hotel type and cancellation status, making it easy to identify which hotel category experiences more booking cancellations.Answer Here.

##### 2. What is/are the insight(s) found from the chart?

• City Hotels have a higher number of total bookings compared to Resort Hotels.

• The number of cancelled bookings is considerably higher for City Hotels than Resort Hotels.

• Both hotel types have more successful (non-cancelled) bookings than cancelled bookings.

• The larger number of cancellations in City Hotels may be associated with business travel, flexible booking policies, or frequent changes in travel plans.

• Resort Hotels show relatively fewer cancellations, suggesting that leisure travelers may be more likely to keep their reservations.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. The analysis helps hotel management identify where cancellation rates are higher and implement strategies to reduce revenue loss.

Positive Business Impact:
• City Hotels can reduce cancellations by introducing stricter cancellation policies, confirmation reminders, or partial advance payments.
• Resort Hotels can continue promoting vacation packages and customer loyalty programs to maintain their comparatively lower cancellation rate.
• Understanding cancellation patterns helps improve occupancy planning, pricing strategies, and resource allocation.

Potential Negative Impact:
A high cancellation rate in City Hotels can lead to revenue loss, inefficient room allocation, and increased operational uncertainty. Addressing these issues can improve overall business performance.

#### Chart - 3

````python
plt.figure(figsize=(12,6))
order = ['January','February','March','April','May','June',
         'July','August','September','October','November','December']
sns.countplot(
    data=df1,
    x='arrival_date_month',
    order=order,
    palette='mako'
)
plt.title(
    'Monthly Hotel Booking Trends',
    fontsize=16,
    fontweight='bold'
)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Bookings', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()
````

##### 1. Why did you pick the specific chart?

A count plot was chosen because the variable 'arrival_date_month' is categorical and represents the month in which bookings were made. This chart effectively visualizes the distribution of bookings across different months and helps identify seasonal booking trends and peak travel periods.

##### 2. What is/are the insight(s) found from the chart?

• Hotel bookings gradually increase from January to May.

• Bookings rise sharply during July and reach their highest level in August, making it the peak booking month.

• After August, bookings decline noticeably from September onwards.

• November and December have comparatively fewer bookings, indicating lower travel demand during these months.

• The chart clearly demonstrates a seasonal pattern in hotel reservations, with the highest demand occurring during the summer vacation period.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. Identifying seasonal booking patterns helps hotel management make better business decisions.

Positive Business Impact:
• Hotels can increase room prices during peak months such as July and August to maximize revenue.
• Additional staff and hotel resources can be planned during high-demand seasons.
• Marketing campaigns and promotional offers can be introduced during low-demand months like November and January to improve occupancy.
• Seasonal demand forecasting helps optimize inventory management and operational planning.

Potential Negative Impact:
Failure to prepare for peak seasons may lead to overbooking, poor customer service, and operational challenges. Similarly, low-demand months may result in reduced revenue if promotional strategies are not implemented.

#### Chart - 4

````python
plt.figure(figsize=(10,6))
order = df1['market_segment'].value_counts().index
sns.countplot(
    data=df1,
    y='market_segment',
    order=order,
    palette='rocket'
)
plt.title(
    'Bookings by Market Segment',
    fontsize=16,
    fontweight='bold'
)
plt.xlabel('Number of Bookings', fontsize=12)
plt.ylabel('Market Segment', fontsize=12)
plt.grid(axis='x', alpha=0.3)
plt.show()
````

##### 1. Why did you pick the specific chart?

A horizontal count plot was chosen because 'market_segment' is a categorical variable with multiple categories. This chart makes it easy to compare the number of bookings across different market segments and identify which customer acquisition channels contribute the most to hotel bookings.

##### 2. What is/are the insight(s) found from the chart?

• Online TA (Online Travel Agencies) is the largest source of hotel bookings by a significant margin.

• Offline TA/TO is the second-largest booking channel, followed by Direct bookings.

• Corporate and Group bookings contribute a moderate number of reservations.

• Complimentary, Aviation, and Undefined market segments contribute only a very small percentage of total bookings.

• The hotel business is highly dependent on online booking platforms, indicating the growing importance of digital travel agencies.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. This analysis helps hotel management understand which booking channels generate the highest customer traffic and where marketing efforts should be focused.

Positive Business Impact:
• Hotels can strengthen partnerships with Online Travel Agencies to maximize bookings.
• Marketing budgets can be allocated more efficiently by investing in high-performing booking channels.
• Direct booking campaigns can be improved to reduce dependency on third-party platforms and increase profit margins.
• Underperforming market segments can be evaluated to determine whether additional promotional strategies are required.

Potential Negative Impact:
Heavy dependence on Online Travel Agencies may increase commission costs and reduce profit margins. Hotels should encourage more direct bookings through loyalty programs, discounts, and exclusive website offers.

#### Chart - 5

````python
plt.figure(figsize=(10,6))
sns.boxplot(
    data=df1,
    x='is_canceled',
    y='lead_time',
    palette='crest'
)
plt.title(
    'Lead Time vs Booking Cancellation',
    fontsize=16,
    fontweight='bold'
)
plt.xlabel('Booking Cancelled (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Lead Time (Days)', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
````

##### 1. Why did you pick the specific chart?

A box plot was selected to compare the distribution of lead time for cancelled and non-cancelled bookings. It helps visualize the median, spread, variability, and presence of outliers, making it suitable for comparing booking behavior between the two groups.

##### 2. What is/are the insight(s) found from the chart?

• Customers who cancelled their bookings generally have a higher lead time than customers who completed their stay.

• The median lead time for cancelled bookings is considerably higher than that of non-cancelled bookings.

• Both groups contain several outliers, indicating that some customers booked several hundred days in advance.

• The wider spread of the cancelled bookings suggests greater variability in booking behavior among customers who cancel.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. Hotels can identify high-risk bookings based on lead time and implement suitable cancellation policies.

Positive Business Impact:
• Hotels can request advance payments for bookings made far in advance.
• Reminder notifications can be sent to customers with long lead times.
• Revenue loss due to cancellations can be reduced through better overbooking and pricing strategies.

Potential Negative Impact:
Customers with very high lead times are more likely to cancel, which may lead to vacant rooms and reduced revenue if appropriate cancellation management strategies are not implemented.
