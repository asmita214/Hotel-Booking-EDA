## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df1 = df.copy()
df1.drop_duplicates(inplace=True)
df1['children'] = df1['children'].fillna(df1['children'].median())
df1['country'] = df1['country'].fillna(df1['country'].mode()[0])
df1['agent'] = df1['agent'].fillna(0)
df1['company'] = df1['company'].fillna(0)
print("Dataset Shape :", df1.shape)
print("\nMissing Values:\n")
print(df1.isnull().sum())
````

### What all manipulations have you done and insights you found?

### Data Manipulations Performed

• Loaded the Hotel Booking Demand dataset into a Pandas DataFrame.
• Explored the dataset using head(), shape(), info(), describe(), and columns.
• Identified duplicate records and removed 31,994 duplicate rows.
• Detected missing values in the columns company, agent, country, and children.
• Filled missing values:
   - children → Median
   - country → Mode
   - agent → 0
   - company → 0
• Verified that the cleaned dataset contained no missing values.
• Checked the unique values for every variable to understand categorical and numerical features.

### Key Insights

• Final cleaned dataset contains 87,396 observations and 32 features.
• The dataset includes both categorical and numerical variables, making it suitable for EDA.
• Hotel bookings span three years (2015–2017).
• Variables such as ADR, Lead Time, and Previous Bookings show large variability, indicating possible outliers.
• Binary variables like is_canceled and is_repeated_guest are useful for comparison-based analysis.
